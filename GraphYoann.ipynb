```
import pandas as pd
import pylab
```

```
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("worldplayers.csv")
```

```
df.loc[(df['Gls'])>18]
```

```
df.loc[ (df['Age']<=20) & (df['Gls']>=10)]
```

```
R1=df.loc[(df['Gls']>18)&(df['Ast']>4)]['name']
```

```
L1=[]
for elt in R1:
  L1.append(elt)
```

```
R2=df.loc[(df['Gls']>18)&(df['Ast']>4)]['Gls']
```

```
L2=[]
for elt in R2:
  L2.append(elt)
```

```
fig = plt.figure()
barName=[]
x = L1
height = L2
width = 0.2

plt.bar(x, height, width, color='b' )
plt.title("buts marques par joueurs plus de 18 buts et 4 passes")

plt.show()
```

```
labels = L1
sizes = L2
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.title("buts marques par joueurs plus de 18 buts et 4 passes")
plt.show()
```

```
R3=df.loc[(df['Squad']=='Brest')&(df['Gls']>1)]['name']
```

```
R4=df.loc[(df['Squad']=='Brest')&(df['Gls']>1)]['Gls']
```

```
R5=df.loc[(df['Squad']=='Brest')&(df['Gls']>1)]['Min']
```

```
L3=[]
for elt in R3:
  L3.append(elt)
```

```
L4=[]
for elt in R4:
  L4.append(elt)
```

```
L5=[]
for elt in R5:
  elt=str(elt)
  if len(elt)<4:
    elt=int(elt)
    L5.append(elt)
  else:
    elt=elt.replace(',','')
    elt=int(elt)
    L5.append(elt)


  
```

```
labels = L3
sizes = L5
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','green','purple']

plt.pie(sizes, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.title("buts marques par les joueurs du Stade Brestois")
plt.show()
```
