```
from urllib.request import urlopen
from bs4 import BeautifulSoup
from bs4 import Comment
import pandas as pd
from bs4.element import Tag 
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('worldplayers.csv')
#print(df['name'][0])

```

```
R1 = df.loc[(df['Nation'] == 'eng ENG')&(df['division'] =='1')&(df['Pos'].str.contains('FW'))]
R1
```

```
gls = R1.groupby("name")["Gls"].sum().reset_index()
gls = gls.sort_values(by="Gls", ascending = False)
plt.figure(figsize=(13,13))
plt.subplot(122)
ax = sns.barplot(y="name", x ="Gls", data =  gls, palette="summer", linewidth = 1, edgecolor ="k"*20)
plt.ylabel('')
plt.subplots_adjust(wspace= 0.1)
plt.title("Top English forward in goals")
for i,j in enumerate(gls["Gls"][:20]):
  ax.text(0.7,i,j,weight ="bold")
plt.show()
```
